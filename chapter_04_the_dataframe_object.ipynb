## 4.1 The outline of DataFrame

```python
import numpy as np
import pandas as pd
```

```python
city_data = {
    "City" : ["New York", "Paris", "Barcelona", "Rome"],
    "Country" : ["United States", "France", "Spain", "Italy"],
    "Population" : pd.Series([860000, 2141000, 5515000, 2873000])
}

cities = pd.DataFrame(city_data)
cities
```

```python
cities.transpose()
```

```python
cities.T
```

```python
random_data = np.random.randint(1, 101, [3,5])
```

```python
random_data
```

```python
pd.DataFrame(data = random_data)
```

```python
row_labels = ["Morning", "Afternoon", "Evening"]
temperatures = pd.DataFrame(
    data = random_data, index = row_labels
)
temperatures
```

```python
column_labels = (
    "Monday",
    "Tuesday",
    "Wednesday",
    "Thursday",
    "Friday"
)

pd.DataFrame(
    data = random_data, index = row_labels, columns = column_labels
)
```
