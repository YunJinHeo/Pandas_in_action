## 8.2 Create pivot_table in the DataFrame

```python
pd.read_csv("sales_by_employee.csv").head()
```

```python
sales = pd.read_csv("sales_by_employee.csv", parse_dates = ["Date"])
```

```python
sales.head()
```

```python
sales.pivot_table(index = "Date", values = ["Revenue", "Expenses"], aggfunc= "sum")
```

```python
sales.pivot_table(
    index = "Date",
    columns = "Name",
    values = "Revenue",
    aggfunc = "sum"
)
```

```python
sales.pivot_table(
    index = "Date",
    columns = "Name",
    values = "Revenue",
    aggfunc = "sum",
    fill_value = 0
)
```

```python
sales.pivot_table(
    index = "Date",
    columns = "Name",
    values = "Revenue",
    aggfunc = "sum",
    fill_value = 0,
    margins = True
)
```

```python
sales.pivot_table(
    index = "Date",
    columns = "Name",
    values = "Revenue",
    aggfunc = "sum",
    fill_value = 0,
    margins = True,
    margins_name = "Total"
)
```

```python
sales.pivot_table(
    index = "Date",
    columns = "Name",
    values = "Revenue",
    aggfunc = "count"
)
```

```python
sales.pivot_table(
    index = "Date",
    columns = "Name",
    values = "Revenue",
    aggfunc = ["sum", "count"]
)
```

```python
sales.pivot_table(
    index = "Date",
    columns = "Name",
    values = ["Revenue", "Expenses"],
    aggfunc = {"Revenue" : "min", "Expenses" : "min"}
)
```

```python
sales.pivot_table(
    index = ["Name", "Date"],
    values = "Revenue",
    aggfunc = "sum"
).head()
```

```python
sales.pivot_table(
    index = ["Date", "Name"],
    values = "Revenue",
    aggfunc = "sum"
).head(10)
```

## 8.3 Index level stack and unstack

```python
sales.head()
```

```python
by_name_and_date = sales.pivot_table(
    index = "Name",
    columns = "Date",
    values = "Revenue",
    aggfunc = "sum"
)
```

```python
by_name_and_date
```

```python
by_name_and_date.stack()
```

```python
sales_by_customer = sales.pivot_table(
    index = ["Customer", "Name"],
    values = "Revenue",
    aggfunc = "sum"
)
sales_by_customer.head()
```

```python
sales_by_customer.unstack()
```

```python
video_game_sales = pd.read_csv("video_game_sales.csv")
```

```python
video_game_sales
```

```python
video_game_sales.melt(id_vars = "Name", value_vars = "NA").head()
```

```python
video_game_sales.melt(id_vars = "Name", value_vars = ["NA", "EU", "JP", "Other"])
```

```python
video_game_sales_by_region = video_game_sales.melt(id_vars = "Name", value_vars = ["NA", "EU", "JP", "Other"], var_name = "Region", value_name = "Sales")
```

```python
video_game_sales_by_region
```

```python
video_game_sales_by_region.pivot_table(
    index = "Name",
    values = "Sales",
    aggfunc = "sum"
).sort_values("Sales", ascending = False)
```
