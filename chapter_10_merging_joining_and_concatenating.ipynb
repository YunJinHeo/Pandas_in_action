## 10.6 Outer join

```python
groups.head()
```

```python
cities.head()
```

```python
groups.merge(cities, how = 'outer', left_on = 'city_id', right_on = 'id')
```

```python
groups.merge(
    cities,
    how = 'outer',
    left_on = 'city_id',
    right_on = 'id',
    indicator = True
)
```

```python
outer_join = groups.merge(
    cities,
    how = 'outer',
    left_on = 'city_id',
    right_on = 'id',
    indicator = True
)
in_right_only = outer_join['_merge'] == 'right_only'

outer_join[in_right_only].head()
```
